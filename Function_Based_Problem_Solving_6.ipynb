```python
def parking_system(a, b):
    x = 0   
    for i in b:
        if i == "in":
            x = x + 1
        elif i == "out":
            if x > 0:
                x = x - 1
    print("Currently Parked Vehicles:", x)
    if x > a:
        print("Parking Status: Exceeded Capacity")
    else:
        print("Parking Status: Available")
a = int(input("Parking Capacity: "))
b = input("Vehicle Logs (separate by space): ").split()
parking_system(a, b)
```

```python
def event_controller(a, b):
    x = 0   
    y = 0  
    for i in range(b):
        if x < a:
            x = x + 1
        else:
            y = y + 1
    print("Confirmed Registrations:",x)
    print("Waitlisted Users:",y)
    if x == a:
        print("Registration Status:Closed")
    else:
        print("Registration Status:Open")
a = int(input("Event Capacity:"))
b = int(input("Registrations:"))
event_controller(a,b)
```

```python
def delivery_time(a, b, c):
    x = a * 5   
    if b.lower() == "high":
        x = x + 10
    elif b.lower() == "medium":
        x = x + 5
    if c.lower() == "rainy":
        x = x + 5
    elif c.lower() == "stormy":
        x = x + 10
    print("Estimated Delivery Time:",x,"minutes")
a = int(input("Distance (km):"))
b = input("Traffic Level:")
c = input("Weather:")
delivery_time(a,b,c)
```
