```python
a = {"Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85}
b = 0
for x in a:
    if a[x] > b:
        b = a[x]
print("Top Performers Eligible for Bonus:")
for x in a:
    if a[x] == b:
        print(x, b)
```

```python
a = [3, 4, 7, 8, 10, 12, 5]
b = [] 
for x in range(len(a)):
    if a[x] % 2 == 0:     
        b.append((x, a[x]))
print("Valid Sensor Readings (Hour, Value):", b)
```

```python
a = ["ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"]
b = {}  
for x in a:    
    c = x.split
    if c in b:
        b[c] = b[c] + 1
    else:
        b[c] = 1
```
