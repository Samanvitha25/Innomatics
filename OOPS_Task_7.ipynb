```python
class Employee:
    def __init__(self, employee_name, employee_id, department):
        self.employee_name = employee_name
        self.employee_id = employee_id
        self.department = department
    def show_id_card(self):
        print("Employee ID Card")
        print("Name:", self.employee_name)
        print("ID:", self.employee_id)
        print("Department:", self.department)
employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"
employee.show_id_card()
```

```python
class Contact:
    def __init__(self, contact_name, phone_number):
        self.contact_name = contact_name
        self.phone_number = phone_number
    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.contact_name)
        print("Phone:", self.phone_number)
contact_name = "Anita"
phone_number = "9876543210"
contact = Contact(contact_name, phone_number)
contact.show_contact()
```

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    def show_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))
product_name = "Headphones"
price = 2499
product = Product(product_name, price)
product.show_price_tag()
```

```python
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    def show_details(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")
movie_name = "Inception"
rating = 4.8
movie = Movie(movie_name, rating)
movie.show_details()
```
