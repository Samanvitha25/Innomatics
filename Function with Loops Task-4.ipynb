```python
def validate_recharge(recharge_amount):
    valid_plans = [199, 299, 399, 599]
    if recharge_amount < 50:
        print("Recharge amount must be at least 50")
        return False
    elif recharge_amount not in valid_plans:
        print("Invalid plan. Choose from 199, 299, 399, 599")
        return False
    else:
        print("Recharge Successful")
        return True
while True:
    user_amount = int(input("Enter recharge amount:"))
    if validate_recharge(user_amount):
        break
    else:
        print("Please try again")
```

```python
def check_inventory(products_stock):
    for product in products_stock:
        stock = products_stock[product]       
        if stock < 15:
            result = "Reorder Alert"
        else:
            result = "Stock OK"
        print(product, "-", result)
products_stock = {}
n = int(input("Enter number of products: "))
for i in range(n):
    product = input("Enter product name: ")
    stock = int(input("Enter stock quantity: "))
    products_stock[product] = stock
print("\nInventory Status:")
check_inventory(products_stock)
```

```python
def check_result(student_marks):
    total = 0
    for marks in student_marks:
        total = total + marks
    average = total / len(student_marks)
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    return result
student_marks = []
n = int(input("Enter number of subjects: "))
for i in range(n):
    marks = int(input("Enter marks: "))
    student_marks.append(marks)
final_result = check_result(student_marks)
print("Result:", final_result)
```

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km = 12
    total_fare = base_fare + (distance * per_km)
    if peak_hour == "yes":
        total_fare = total_fare + (total_fare * 0.25)
    return total_fare
while True:
    distance = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")
    fare = calculate_fare(distance, peak_hour)
    print("Total Fare: ₹", fare)
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry != "yes":
        break
```
