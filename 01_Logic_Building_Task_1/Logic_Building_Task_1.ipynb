```python
x = "admin"
y = "1234"
username = input("Enter username:")
password = input("Enter password:")
if username == x and password == y:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]
x = 0
y = 0
for mark in marks:
    if mark >=50:
        x += 1
    else:
        y += 1
print("Pass Students:",x)
print("Fail Students:",y)
```

```python
x = [" Alice ", "bob", " CHARLIE "]
y = []
for a in x:
    b = a.strip()  
    b = b.lower()
    y.append(b)
print("Cleaned List:",y)
```

```python
x = ["Hi", "Welcome to the platform", "OK"]
for a in x:
    b = len(a)
    print("Message:", a)
    print("Length:",b) 
    if b > 10:
        print("Status: Long Message")
    else:
        print("Status:Short Message")  
    print()
```

```python
x = ["INFO", "ERROR", "WARNING", "ERROR"]
y = 0
for a in x:
    if a == "ERROR":
        y += 1
print("Total ERROR:",y)
```
