```python
def analyze_post(a):
    b = 0
    for x in a:
        b = b + x
    if b >= 1000:
        c = "Viral Post"
    else:
        c = "Normal Engagement"
    print("Total Likes:", b)
    print("Post Status:", c)
a = []
n = int(input("Enter number of likes: "))
for i in range(n):
    x = int(input("Enter like: "))
    a.append(x)
analyze_post(a)
```

```python
def check_stock(a):
    if a < 10:
        print("Low Stock Alert")
    else:
        print("Stock Sufficient")
a = int(input("Enter medicine stock:"))
check_stock(a)
```

```python
def check_duplicates(a):
    b = set(a)
    if len(a) != len(b):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
a = []
n = int(input("Enter number of usernames: "))
for i in range(n):
    x = input("Enter username: ")
    a.append(x)
check_duplicates(a)
```

```python
def check_eligibility(a):
    if a >= 18:
        b = "Eligible"
    else:
        b = "Not Eligible"   
    print("Patient Age:", a)
    print("Eligibility Status:", b)
a = int(input("Enter patient age: "))
check_eligibility(a)
```

```python
def filter_premium(a):
    b = []
    for x in a:
        if x > 2000:
            b.append(x)
    print("Premium Crops:", b)
a = []
n = int(input("Enter number of crop prices: "))
for i in range(n):
    x = int(input("Enter crop price: "))
    a.append(x)
filter_premium(a)
```

```python
def check_health(a):
    if a == 0:
        b = "Healthy"
    elif a <= 5:
        b = "Minor Issues"
    else:
        b = "Critical Issues"
    print("Error Count:", a)
    print("System Status:", b)
a = int(input("Enter error count: "))
check_health(a)
```

```python
def check_transaction(a):
    b = 50000
    if a <= b:
        c = "Approved"
    else:
        c = "Rejected"
    print("Transaction Amount:", a)
    print("Transaction Status:", c)
a = int(input("Enter transaction amount: "))
check_transaction(a)
```
